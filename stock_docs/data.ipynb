```python
import tushare as ts
import datetime
import time
import os
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
curr_date = datetime.datetime.now().strftime('%Y%m%d')
daily_dir = 'D:/share_data/daily/' + curr_date + '/'
if not os.path.exists(daily_dir):
    os.mkdir(daily_dir)

stock_daily = ts.get_today_all()
stock_daily.to_csv(daily_dir + 'daily.csv')
```

```python
stock_info = ts.get_stock_basics()

def get_min_detail_data(min_dir,ktype,curr_date,finish_num=0):
    if not os.path.exists(min_dir):
        os.mkdir(min_dir)
    
    start_date = curr_date
    idx = 0
    for code in stock_info.index:
        idx += 1
        if idx < finish_num:
            continue
        if idx % 100 == 0:
            print('curr idx: ' + str(idx))        

        data_min = ts.get_hist_data(code, start = start_date, ktype = ktype)
        if data_min is None:
            #print('empty data : ' + code)
            continue
        
        data_min.to_csv(min_dir + code + '.csv')
```

```python
curr_date = datetime.datetime.now().strftime('%Y-%m-%d')
#curr_date = '2019-09-04'
get_min_detail_data(daily_dir + 'day/', 'D',curr_date)
get_min_detail_data(daily_dir + '5min/', '5',curr_date)
get_min_detail_data(daily_dir + '15min/', '15',curr_date)
get_min_detail_data(daily_dir + '30min/', '30',curr_date)
```

```python
def get_tick_data(tick_dir , curr_date, finish_num=0):
    if not os.path.exists(tick_dir):
        os.mkdir(tick_dir)
        
    start_date = curr_date
    idx = 0
    for code in stock_info.index:
        idx += 1
        if idx < finish_num:
            continue
        if idx % 100 == 0:
            print('curr idx: ' + str(idx))

        data_tick = ts.get_tick_data(code,date=start_date,src='tt')
        if data_tick is None:
            #print('empty data : ' + code)
            continue
        
        data_tick.to_csv(tick_dir + code + '.csv')
```

```python
tick_dir = 'D:/share_data/daily/' + curr_date + '_tick/'
#curr_date = '2019-09-04'
get_tick_data(tick_dir,curr_date)
```
